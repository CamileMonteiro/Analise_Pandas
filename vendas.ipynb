# Análise com pandas

```python
import pandas as pd
```

```python
vendas_df = pd.read_excel('dataset.xlsx')
display(vendas_df)
```

## Quais colunas possuem valores ausentes?
Nenhuma coluna deu valor nulo.

```python
print(vendas_df.isnull().sum())
```

## Qual é o valor médio de “Total_Vendas” e “Lucro”?
-246.49 | -28.61

```python
display(vendas_df.describe())
```

## Qual é a região com maior volume de vendas?

```python
regiao_maior = vendas_df.groupby("Regiao")["Total_Vendas"].sum().sort_values(ascending=False).head(5)
display(regiao_maior)
```

## Qual é a região com menor volume de vendas?

```python
regiao_menor = vendas_df.groupby("Regiao")["Total_Vendas"].sum().sort_values(ascending=False).tail(5)
display(regiao_menor)
```

## Gráficos

```python
import matplotlib.pyplot as plt
```

```python
vendas_df.groupby("Regiao")["Total_Vendas"].sum().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Regiões com Maior Total de Vendas")
plt.xlabel("Região")
plt.ylabel("Total de Vendas")
plt.show()
```

## Com base no regiao_menor, apenas com (tail(5))

```python
menores = regiao_menor.sort_values(ascending=True)
menores.plot(kind='bar', figsize=(10,5))
plt.title("Regiões com Menor Total de Vendas")
plt.xlabel("Região")
plt.ylabel("Total de Vendas")
plt.show()
```
